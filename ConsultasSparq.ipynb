```python
from SPARQLWrapper import SPARQLWrapper, JSON
from pandas import json_normalize
```

```python
sparql = SPARQLWrapper(
    "http://MacBook-Pro-de-Diego.local:7200/repositories/Hotels"
)
sparql.setReturnFormat(JSON)
```

```python
def get_results(query):
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)

    results = sparql.query().convert()
    return(results)
```

**Pregunta 1**

¿Cual es la temperatura de la ubicacion del sitio turistico de interes?

```python
query = """
    PREFIX schema: <https://schema.org/>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX ns1: <https://bimerr.iot.linkeddata.es/def/weather/>
    PREFIX xsd: <http://www.w3.org/2001/XMLSchema#> 
    PREFIX dbo: <http://dbpedia.org/ontology/>
    PREFIX ns2: <https://bimerr.iot.linkeddata.es/def/weather/>
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>
    select ?nombre ?temperatura
    where { 
	?TouristDestination rdf:type dbo:City .
	?TouristDestination schema:name ?nombre .
    ?TouristDestination ns2:temp ?temperatura .
    ?TouristDestination schema:State ?Provincia .
    ?Provincia foaf:name ?nombreProvincia .
    Filter (xsd:string(?nombreProvincia)="Azuay" && xsd:float(?temperatura) < 294.47)
    }
    """

results_df = json_normalize((get_results(query))['results']['bindings'])[[
    'nombre.value', 'temperatura.value'
]]

results_df
```

```python
query = """
    prefix ns1: <http://www.purl.org/dabgeo/common-domain/weather#/> 
    prefix ns2: <https://bimerr.iot.linkeddata.es/def/weather/> 
    prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> 
    prefix schema: <https://schema.org/> 
    prefix xsd: <http://www.w3.org/2001/XMLSchema#> 
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

    select ?nombre ?Clima
    where { 
	    ?TouristDestination rdf:type schema:TouristDestination .
        ?TouristDestination schema:State ?State .
        ?TouristDestination schema:name ?nombre .
        ?State rdfs:label ?nombreProvincia .
        ?TouristDestination ns1:WeatherCondition ?weather .
        ?weather ns1:hasWeatherCondition ?Clima .
    }
    """

results_df = json_normalize((get_results(query))['results']['bindings'])[[
    'nombre.value', 'Clima.value'
]]

results_df
```

¿Cuáles lugrares turisticos tienen una presión menor a 1014?

```python
query = """
    prefix ns1: <http://www.purl.org/dabgeo/common-domain/weather#/> 
    prefix ns2: <https://bimerr.iot.linkeddata.es/def/weather/> 
    prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> 
    prefix schema: <https://schema.org/> 
    prefix xsd: <http://www.w3.org/2001/XMLSchema#>  
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

    select ?nombre ?presion
    where { 
	    ?TouristDestination rdf:type schema:TouristDestination .
        ?TouristDestination schema:State ?State .
        ?TouristDestination schema:name ?nombre .
        ?State rdfs:label ?nombreProvincia .
        ?TouristDestination ns2:pressure ?presion .
        Filter(xsd:integer(?presion) < 1014)
    }
"""

results_df = json_normalize((get_results(query))['results']['bindings'])[[
    'nombre.value', 'presion.value'
]]

results_df
```

Pregunta

¿Qué lugares turísticos tiene de estado de clima "Rain" en la provincia de "Pichincha"? 

```python
query = """
    prefix ns1: <http://www.purl.org/dabgeo/common-domain/weather#/> 
    prefix ns2: <https://bimerr.iot.linkeddata.es/def/weather/> 
    prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> 
    prefix schema: <https://schema.org/> 
    prefix xsd: <http://www.w3.org/2001/XMLSchema#> 
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
	PREFIX foaf: <http://xmlns.com/foaf/0.1/>

    select ?nombre ?nombreProvincia ?Clima
    where { 
	    ?TouristDestination rdf:type schema:TouristDestination .
        ?TouristDestination schema:State ?State .
        ?TouristDestination schema:name ?nombre .
        ?State foaf:name ?nombreProvincia .
        ?TouristDestination ns1:WeatherState ?weather .
        ?weather ns1:hasWeatherCondition ?Clima .
    Filter(xsd:string(?Clima) = "Rain" && xsd:string(?nombreProvincia) = "Pichincha")
    }
    """

results_df = json_normalize((get_results(query))['results']['bindings'])[[
    'nombre.value', 'Clima.value'
]]

results_df
```

Lugares que pretenecen a provincias con una área menor a 10000000000

```python
query = """
prefix ns1: <http://www.purl.org/dabgeo/common-domain/weather#/> 
prefix ns2: <https://bimerr.iot.linkeddata.es/def/weather/> 
prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> 
prefix schema: <https://schema.org/> 
prefix xsd: <http://www.w3.org/2001/XMLSchema#> 
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX dbo: <http://dbpedia.org/ontology/>

select ?nombre ?nombreProvincia ?Area ?descripcion
where { 	
    ?TouristDestination rdf:type schema:TouristDestination .
    ?TouristDestination schema:State ?State .
    ?TouristDestination schema:name ?nombre .
    ?Provincia rdf:type schema:State .
    ?Provincia dbo:areaTotal ?Area .
    ?Provincia rdfs:label ?nombreProvincia .
    ?Provincia rdfs:comment ?descripcion .
    Filter (xsd:double(?Area) < 10000000000) .
}
"""
results_df = json_normalize((get_results(query))['results']['bindings'])[[
    'nombre.value', 'nombreProvincia.value', 'Area.value', 'descripcion.value'
]]

results_df
```
